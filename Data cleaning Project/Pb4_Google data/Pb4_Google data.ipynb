## Cheaking Rating Column

```python
# Checking null value 
```

```python
df["Rating"].isnull().sum()
```

### Fill nas as blank

```python
df = df.fillna("")
```

```python
### Change Price column as float type 
```

```python
df.Price.unique()
```

```python
# Remve charters in the column
df["Price"] = df["Price"].str.replace("$","")
```

```python
df.Price.unique()
```

```python
# Remove str in the column 
df["Price"] = df["Price"].str.replace("Everyone","0")
```

```python
df.Price.unique()
```

```python
df["Price"].astype(float)
```

```python
df["Price"] = df["Price"].astype(float)
```

```python
df.info()
```

```python
# Renaming the Price column
df = df.rename( columns ={ "Price" : "Price in $"})
```

```python
df
```

### Removing Type column 

```python
df = df.drop(columns = "Type")
```

```python
df
```

### Modify the Rating colum as 1 decimal point

```python
df["Rating"]
```

```python
df["Rating"] = df["Rating"].round(1)
```
