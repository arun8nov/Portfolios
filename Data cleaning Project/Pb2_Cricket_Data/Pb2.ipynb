```python
# Calculas in df
# Averrage crickter's Career length 
df["Career_length"].mean()
```

```python
# Max Highest inns score by country
df.groupby("Player_Country")["Highest_Inns_Score"].max().to_frame("Highinncount").sort_values("Highinncount",ascending=False).reset_index()
```
